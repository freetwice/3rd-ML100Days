# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

# ReLU數學式
def relu(x):
    return list(map(lambda x:x if x>0 else 0, x))
   
# ReLU一階導數
def dev_relu(x):
    return list(map(lambda x:1 if x>0 else 0, x))
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, relu(x), color='b', linewidth='2')
plt.grid()
plt.title('ReLU Function')

plt.text(4,0.8, 'relu(x)=max(0,x)', fontsize=12)

plt.show()
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, dev_relu(x), color='b', linewidth='2')
plt.grid()
plt.title('ReLU dev_Function')

plt.text(4,0.8, 'relu\'(x)', fontsize=15)

plt.show()
```
